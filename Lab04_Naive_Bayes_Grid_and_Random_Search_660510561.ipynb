```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

from scipy.stats import uniform
```

```python
train = fetch_20newsgroups(subset='train')
test = fetch_20newsgroups(subset='test')

Xtrain = train.data[:3000]
ytrain = train.target[:3000]
Xtest = test.data[:500]
ytest = test.target[:500]

print("X:", len(Xtest))
print("y:", len(ytest))
```

### Naive Bayes [(Documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html)

```python
pipeline = Pipeline([('tfidf', TfidfVectorizer(stop_words = 'english')),
                     ('nb', MultinomialNB())])

pipeline.fit(Xtrain, ytrain)

# เมื่อ fit เสร็จ ก็เอาไปทำนายบน test set
ypred = pipeline.predict(Xtest)

print(classification_report(ytest, ypred))
```

### Random Search Cross-Validation [(Documentation)](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html)

### Uniform distribution in `Scipy` [(Documentation)](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html)

```python
# การ search หา พารามิเตอร์ที่เหมาะสม
parameters = {'nb__alpha': uniform(loc=0, scale=10)}

clf = RandomizedSearchCV(pipeline, parameters, n_iter=7)

clf.fit(Xtrain, ytrain) # ผลลัพธ์ที่ได้จะเป็นโมเดลที่ดีที่สุด
```

```python
ypred = clf.predict(Xtest)

print(classification_report(ytest, ypred))
```

ทำ 5-fold ให้ cv = 5

ขั้นตอนสุดท้ายบอกด้วยว่า ระหว่าง grid search และ randomize ตัวไหนให้ค่าพารามิเตอร์ที่ดีที่สุด

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
pipeline_grid = Pipeline([('tfidf', TfidfVectorizer(stop_words='english')),
 ('nb', MultinomialNB())])

parameters_grid = {'nb__alpha': [0.01, 0.1, 0.5, 1.0, 5.0]}

grid = GridSearchCV(pipeline_grid, parameters_grid,
                      scoring='f1_macro', cv=5)

grid.fit(Xtrain, ytrain)

best_alpha_grid = grid.best_params_['nb__alpha']
print("Best alpha (Grid Search):", best_alpha_grid)
```

```python
from sklearn.metrics import f1_score
```

```python
y_pred_grid = grid.predict(Xtest)

f1_grid = f1_score(ytest, y_pred_grid, average='macro')

print(classification_report(ytest, y_pred_grid))
print("f1_macro (Grid Search):", f1_grid)
```

```python
pipeline_random_search = Pipeline([('tfidf', TfidfVectorizer(stop_words='english')),
    ('nb', MultinomialNB())
])

parameters_random_search = {'nb__alpha': uniform(loc=0, scale=10)}

clf_random_search = RandomizedSearchCV(pipeline_random_search,
                                       parameters_random_search, n_iter=15,
                                       scoring='f1_macro', cv=5)

clf_random_search.fit(Xtrain, ytrain)
```

```python
ypred_random_search = clf_random_search.predict(Xtest)

print(classification_report(ytest, ypred_random_search))
```

```python
print("Best alpha (Random Search):", clf_random_search.best_params_['nb__alpha'])
```

```python
ypred_random_search = clf_random_search.predict(Xtest)

print(classification_report(ytest, ypred_random_search))

f1_random_search = f1_score(ytest, ypred_random_search, average='macro')
```

```python
print("f1_macro (Random Search):", f1_random_search)
```
